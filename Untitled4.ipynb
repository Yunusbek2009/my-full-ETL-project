```python
# sending a request based of a link
# 1- importing the library to do that 
import requests
# writing down the url
url = 'https://talk.collegeconfidential.com/c/colleges-and-universities/77/none'
# if my requests reseives a response of 200 that means that scarpping good to go
# and then turning it 
response = requests.get(url)
```

```python
# specifying which libraries that i will be using 
import requests
from bs4 import BeautifulSoup
import re
# turning my messy response into a much better format so that i can infultate it 
soup = BeautifulSoup(response.text , 'html')

# Get all 'description' and 'name' spans
filtered_descriptions = soup.find_all('span', itemprop='description')
filtered_names = soup.find_all('span', itemprop='name')

# Combine them into a dictionary or a desired format
result = []

for name, description in zip(filtered_names, filtered_descriptions):
    result.append({
        'university': name.get_text(),
        'description': description.get_text()
    })

# since i have a good understanding of csv file i can turn my data into csv file and save it correctly
# and then actually saving it to my directory as this just in case it doesn't get deleted or if something goes wrong i will be able to use it again

import csv
with open('mydata /project_data2.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=['university', 'description'])
    # telling my computer that my data contains a header
    writer.writeheader()
    # so do rows
    writer.writerows(result)
```

```python
# i have had my library imported allready so as i am using jupiter notebook i do not have to write it again
#convert my csv into a dataframe i use read_csv method
df = pd.read_csv('mydata /project_data2.csv')
# just to see wheather my data is imported correctly if it is not i can specify a few thing to my dataframe
df
# now i am getting the data i need so now it is time for slicing and dicing
```

```python
#extrancing data from two columns (column = 'university , column = 'description)
df['Name'] = df['description'].str.split('•').str[0]
```

```python
#creating a column called Category use the description column
df['Category'] = df['description'].str.split('•').str[1].str.split(' ').str[2]
```

```python
# now i can get acceptance rate
df['Acceptance'] = df['description'].str.split('•').str[2].str.split(' ').str[3]
```

```python
# i am thinking of creating a coumn called duration of study years
df['Duration_of_study'] = df['description'].str.split('•').str[1].str.split(' ').str[1]
```

```python
# just ginging more readable name 
df.rename(columns = {'university':'University_name' , 'Acceptance' : 'Acceptance_rate'} , inplace = True)
```

```python
# just to make it much better and lucrative
# i will be appling a function here
df['Duration_of_study'] = df['Duration_of_study'].apply(lambda x: str(x) + 's')
```

```python
# specifing a city and sate name into two distinct columns
df['City'] = df['Name'].str.split(',').str[0]
```

```python
df['state'] = df['Name'].str.split(',').str[1]
```

```python
# since my Acceptance_rate column contains % and it is in string format i am thing of creating another one for other porposes
# one for clearifying them by their difficulty level and the othe for visualiztion and creating interactive dashboards
df['temporary'] = df['Acceptance_rate'].str.split('%').str[0].fillna(0).astype(int)
```

```python
# scecifying each university based on their difficulity levels 
 # if the acceptance rate is above 7 it means High | severe
 # if the acceptance rate is between 7 and 20 it means Moderate
 # if the acceptance rate is between 20 and 90 it means low
 # if the acceptance rate is equal to 90 or more it means None | Minimal   
 # to achive that i have simply used the lambda dunction to name them for me
 

df['Difficulty_level']  = df['temporary'].apply(lambda x: 'High | Severe' if x <= 7 else 'Moderate' if x<=20 else 'Low' if x<90 else 'None | Minimal')
```

```python
#  General EDA (Exploratory Data Analysis)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# basic dashboard to find the average acceptance of univercities in each City
avg_acceptance_rate = df.groupby('state').agg({'temporary':'mean'}).reset_index()
sns.barplot(x = avg_acceptance_rate['state'] , y = avg_acceptance_rate['temporary'])
plt.xticks(rotation='vertical')
plt.title('average acceptance rate of universities per state')
plt.ylabel('acceptance rate')
plt.tight_layout() 
# saving my file just incase if i will use it in its current format 
# and to do that there is a built in function for that
plt.savefig('barplot.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Dashboard that shows the distributions of the categories from the Category column
# first way of creating it yet a little long 
distributions = df.groupby('Category').count().rename(columns = {'Name':'Number of Universities'}).filter(items=['Category','Number of Universities']).reset_index()
sns.barplot(x = 'Category' , y = 'Number of Universities' , data = distributions)
plt.title(' Distribution of Public vs. Private universities.')
plt.tight_layout() 
plt.savefig('barplot.png', dpi=200, bbox_inches='tight')
plt.show()
# simple yet effective way to that using countplot
sns.countplot(x = df['Category'] )#, hue = df['Category'])
plt.title('Distribution of Public vs. Private universities.')
plt.ylabel('Number of Universities')
plt.tight_layout()
# saving my file just incase if i will use it in its current format 
plt.savefig('countplot.png', dpi=200, bbox_inches='tight')
plt.show()
# it is clear form the chart that private universities are more famous that public ones interms of quantity by almost two folds
```

```python
# now i will create a boxplot to see the acceptance rate for both private and public universities 
sns.boxplot(x = 'Category' , y = 'temporary' , data = df)
plt.title('Base acceptance rate for both private and public Universities')
plt.ylabel('Acceptance rate')
plt.savefig('boxplot.png', dpi=200, bbox_inches='tight')
plt.show()
# overall it can be seen that staticticly entering public universities and collages are much easier that it is with private
# just a sum up
# pulic
    # Median Acceptance Rate: Around 75%
    # IQR: About 50% to 89%
    # Whiskers: Extend roughly from 10% to 100%
# Private
    # Median Acceptance Rate: Around 60%
    # IQR: About 35% to 78%
    # Whiskers: Extend roughly from 5% to 100%
```
